# Plotly Pie and Donut Charts - Visualise Categorical Data: Content Ratings

All Android apps have a content rating like “Everyone” or “Teen” or “Mature 17+”. Let’s take a look at the distribution of the content ratings in our dataset and see how to visualise it with plotly - a popular data visualisation library that you can use alongside or instead of Matplotlib.

First, we’ll count the number of occurrences of each rating with .value_counts()

```python
ratings = df_apps_clean['Content_Rating'].value_counts()
ratings
```

The first step in creating charts with plotly is to import plotly.express. This is the fastest way to create a beautiful graphic with a minimal amount of code in plotly.

```python
import plotly.express as px
```

To create a pie chart we simply call px.pie() and then .show() the resulting figure. Plotly refers to all their figures, be they line charts, bar charts, or pie charts as graph_objects.

```python
fig = px.pie(labels=ratings.index, values=ratings.values)
fig.show()
```

Let’s customise our pie chart. Looking at the .pie() [documentation](https://https://plotly.com/python-api-reference/generated/plotly.express.pie.html) we see a number of parameters that we can set, like title or names.

If you’d like to configure other aspects of the chart, that you can’t see in the list of parameters, you can call a method called .update_traces(). In plotly lingo, “traces” refer to graphical marks on a figure. Think of “traces” as collections of attributes. Here we update the traces to change how the text is displayed.

```python
fig = px.pie(labels=ratings.index, values=ratings.values, title='Content Rating', names=ratings.index)
fig.update_traces(textposition='outside', textinfo='percent+label')
fig.show()
```

To create a donut 🍩 chart, we can simply add a value for the hole argument:

```python
fig = px.pie(labels=ratings.index, values=ratings.values, title='Content Rating', names=ratings.index, hole=0.6)
fig.update_traces(textposition='outside', textinfo='percent+label', textfont_size=15)
fig.show()
```

There is a different way you can create the pie chart, without having to create a new Series like we did when using .value_counts(). You can directly put the df as an argument with the column name as value for the 'names' key and it will automatically group the rows with the same value, like so.

```python
fig = px.pie(data_frame=df_apps_clean, names='Content_Rating', title='Content Rating', hole=0.6)
fig.update_traces(textposition='outside', textinfo='percent+label', textfont_size=15)
fig.show()
```

# Extracting Nested Data from a Column

**Challenge**: How many different types of genres are there? Can an app belong to more than one genre? Check what happens when you use .value_counts() on a column with nested values? See if you can work around this problem by using the .split() function and the DataFrame's [.stack() method](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.stack.html).


If we look at the number of unique values in the Genres column we get 114. But this is not accurate if we have nested data like we do here. We can see this using .value_counts() and looking at the values that just have a single entry. There we see that the semi-colon (;) separates the genre names.

```python
len(df_apps_clean['Genres'].unique())
```

```python
df_apps_clean['Genres'].value_counts().sort_values()[:5]
```

<img src="https://img-c.udemycdn.com/redactor/raw/2020-10-11_13-18-53-e5af936a84354a54fc2ef2d1eea24e77.png">

We somehow need to separate the genre names to get a clear picture. This is where the string’s .split() method comes in handy.

```python
stack = df_apps_clean.Genres.str.split(';')
stack.sample(30, random_state=30)
```

That's not great either, is it? The text was separated (now each Genre is an array of 1 or 2 values, see for example index 7690, where 'Casual' and 'Education' are separated by a coma), but they are still in the same column, so we are in the same situation. However, split() has an optional argument expand that when it's set to True, instead of creating an array in the same column, will create new columns, and thus returns a dataframe.

```python
stack = df_apps_clean.Genres.str.split(';', expand=True)
stack.sample(30, random_state=30)
```

So now we're getting somewhere.. We have 2 columns (that are called "0" and "1" but you can think of them as "Main Genre" and "Additional Genre".. Notice how some values have None there (they already had 1 segment) but the ones that had 2 segments, now have 1 in each column.

Cool, but how do we count them?... One option would be call .value_counts() in the column "0", then do the same for column "1", and add up the values somehow... But there's a better option.

**Step 2 - STACK:** Wouldn't it be easier to count if we could put column "1" below column "0"? In other words, if we could stack them? Surprise, surprise, that's exactly what the stack() function does!

```python
stack = df_apps_clean.Genres.str.split(';', expand=True).stack()
stack.sample(30)
```

Notice how .stack() gave us back a Series from a DataFrame, since it took 2 columns and made it one. **ALSO NOTE** where we see values "0" and "1" that's NOT a column! The 0 and 1 values are INNER-indexes, as mentioned before, we're dealing with a Series now.

```python
stack.index
```

Since stack it's a Series, we can FINALLY call the .value_counts() on it and get the count by individual Genre.

```python
num_genres = stack.value_counts()
num_genres
```

If we compare this with the first count we notice how every row now has only ONE Genre, for instance "Casual" doesn't have the ";Education" part anymore. Also, we can se now that there are 53 genres.

```python
len(num_genres)
```
